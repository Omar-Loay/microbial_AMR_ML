# Identify alleles contributing to AMR by machine learning.
1. Use an ensemble Support Vector Machine (**SVM**), equipped with an **L1-norm** cost function (for feature selection, also biological basis in that relatively few alleles contribute to an AMR phenotype) and **stochastic gradient descent** (for collinearity, subpopulation, randomness).
2. Output figures describing the ensemble of SVM hyperplanes.
    - Figure of alleles weightings across the hyperplanes. 
    - Figure of correlation matrix of the weighting alleles for an SVM ensemble.
    - Dataframes for the above descriptions.

### Load _M. tuberculosis_ pan-genome dataset

```python
import pandas as pd
from tqdm import tqdm
from svm_epistasis_funcs import *

pd.options.mode.chained_assignment = None  # default='warn'

%matplotlib inline
```

```python
### Define locations and filenames for the pan-genome matrices.
DATA_DIR = "../data/" 
SUPP_DIR = "../supplement/"
FINAL_SUPP_DIR = SUPP_DIR+"supplement_final/"
SUPP_SVM_PLANES = FINAL_SUPP_DIR+"Supplementary_Data_File_3/"
SUPP_SVM_DATA = SUPP_DIR + "svm_ensemble_data/"

# load key data matrices
CLUSTER_INFO_FILE = "cluster_info.csv"
STRAIN_INFO_FILE = "PATRIC_genome_info.csv"
X_PANGEN_CLUSTERS_FILE = "pangen_cluster_df.csv"  # the species-cluster (each cluster represents a "gene")
X_PANGEN_ALLELES_FILE = "pangen_allele_df.csv"  # the species-allele matrix
# Y_PHENOTYPES_FILE = "y_drug_TB.csv"  # the species-phenotype matrix
Y_PHENOTYPES_FILE = "resistance_data.csv"  # the species-phenotype matrix
```

```python
# Load resistance data
resistance_data = pd.read_csv(DATA_DIR+Y_PHENOTYPES_FILE, index_col = 0) #resistance_data
drug_list = list(resistance_data.columns.map(str))

# Load pan-genome in both cluster and allele form
pan_cluster_df = pd.read_csv(DATA_DIR+X_PANGEN_CLUSTERS_FILE, index_col = 0) #pan_cluster_df
pan_variant_df = pd.read_csv(DATA_DIR+X_PANGEN_ALLELES_FILE, index_col = 0) #clust_variant_df

# Load cluster information file
clust_info_df = pd.read_csv(DATA_DIR+CLUSTER_INFO_FILE, index_col = 0) #test_df_final

# Load strain information file
strain_info_df = pd.read_csv(DATA_DIR+STRAIN_INFO_FILE, index_col = 0) #....
```

```python
# define these dictionaries for ease of mapping...

# clust_to_rv_fromJson = load_json_obj(DATA_DIR+"clust_to_rv.json")
clust_to_rv = pd.Series(clust_info_df.clust_to_rv.values,index=clust_info_df.index).to_dict()
rv_to_clust = pd.Series(clust_info_df.index,index=clust_info_df.clust_to_rv.values).to_dict()
rv_to_name = pd.Series(clust_info_df.gene_name.values,index=clust_info_df.clust_to_rv.values).to_dict()
clust_to_prod = pd.Series(clust_info_df["product"].values,index=clust_info_df.index).to_dict()
```

### Filter out Mobile elements, PE family proteins, and hypotheticals.
Should drop the total number of 15367 variants (columns) to 9975

```python
pan_variant_df = pan_variant_df.fillna(0).copy()

remove_cols = []
for col in tqdm(pan_variant_df.columns):
    col_name = col.split("_")[0].replace("ter", "ter ")
    clust_to_RV = clust_info_df.loc[col_name, "clust_to_rv"]
    gene_name = clust_info_df.loc[col_name, "gene_name"]
    product = clust_info_df.loc[col_name, "product"]
    
    if clust_to_RV == "0" and gene_name == "0":
        remove_cols.append(col)
    if product == "Mobile element protein" or product == "hypothetical protein":
        remove_cols.append(col)
    if "PE-PGRS" in product or "PPE family" in product or "PE family protein" in product:
        remove_cols.append(col)
    if "hypothetical protein" in product or "Hypothetical protein" in product:
        remove_cols.append(col)
    if "Transposase" in product:
        remove_cols.append(col)

remove_cols = list(set(remove_cols))
print "Number of alleles (variants) before filtering...", len(pan_variant_df.columns)
pan_variant_df.drop(remove_cols, axis=1, inplace=True)
print "Number of alleles after filtering...",len(pan_variant_df.columns)
print "Number of alleles removed...",len(remove_cols)
data_to_plot = pan_variant_df.copy()
```

### Hyperparameterization through sklearn GRIDSEARCH_CV

- Feeds into dictionary below.

```python
# initialize value for the threshold feature selection parameter.
threshold_val = 130

drug_to_threshold_dict = {}

for drug_name in ["ethambutol"]: # "ethambutol", "pyrazinamide", "isoniazid", "pyrazinamide", "4-aminosalicylic_acid", "ethionamide"
    
    if drug_name == "MDR" or drug_name == "XDR":
        X1, y1 = get_MDR_training(resistance_data, data_to_plot, drug_name)
    else:
        y1, X1 = get_target_data(data_to_plot, resistance_data, drug_name)
        X1 = remove_key_antibiotic_clusters(X1, drug_name)
    
    print "Before removing low variance: ", X1.shape
    selector = VarianceThreshold(threshold=.01)
    selector.fit_transform(X1)
    X1 = X1[X1.columns[selector.get_support()]].copy()
    print "After removing low variance: ", X1.shape

    # cross-validation sets
    # cv = StratifiedKFold(n_splits = 10)
    cv = RepeatedStratifiedKFold(n_splits=5,n_repeats=2)

    # estimator
    clf = SGDClassifier(loss="hinge", penalty="l1", shuffle=True, class_weight="balanced", alpha=0.0001) 
    # clf = LogisticRegression(penalty = "l1", class_weight="balanced")
    # feature selection w/ parameter "threshold"
    feat_selection = SelectFromModel(clf, threshold=threshold_val)
    # feat_selection = SelectKBest(clf, threshold=threshold_val)

    # Setup pipeline for SGD-SVM feature selection
    model = Pipeline([
              ('fs', feat_selection),  # SelectFromModel using SGD-SVM
              ('clf', clf), 
            ])

    # Set the parameters for optimizing cross-validation 
    params = {
        'fs__threshold': np.linspace(120, 200, num=20) # try a range from 15 -> 150, 
        # np.linspace(.4, 1, num=10) for logistic regression
    }


    grid = GridSearchCV( model , params, cv=cv)
    grid.fit(X1, y1)
    
    print("The best parameters are %s with a score of %0.2f"
          % (grid.best_params_, grid.best_score_))
    
    drug_to_threshold_dict.update({drug_name: {"fs__threshold": grid.best_params_.values()[0], "best_score": grid.best_score_}})
```

# Allele AMR simulations
1. Set threshold dict for the drugs from the hyperparameterization tests above.
2. Run the cell below, defining the following parameters..
    - **top_num**: number of highly associated alleles to consider for downstream analysis.
    - **training_size_percents**: For each SVM hyperplane, the percentage of the total dataset to sample.
    - **simulation_iterations**: The number of hyperplanes in each ensemble. Distribution converges before 200, so 400 is robust.

```python
threshold_range_dict = {
    "ethambutol": [.7, .9], # 160, 200 for .8 sample size, [120, 140] for maybe .3?
    "pyrazinamide": [120, 140], # [160, 170], train_size = 0.6, [175, 180] for train_size = 0.3
    "isoniazid": [160, 200],
    "ethionamide": [190, 215],
    "ofloxacin": [190, 220],
    "4-aminosalicylic_acid": [160, 210],
    "rifampicin": [160, 200],
    "MDR": [195, 210],
    "XDR": [240, 260],
}
```

```python
threshold_value_dict = {
    
    "XDR": ["SVM", "SGD", 230],
    "MDR": ["SVM", "SGD", 200],

    "isoniazid": ["SVM","SGD", 170], # 170
    "rifampicin": ["SVM","SGD", 170], # 170
    "ethambutol": ["SVM","SGD", 185],
    "pyrazinamide": ["SVM","SGD", 150],
    
    "ofloxacin": ["SVM","SGD", 180],
    "streptomycin": ["SVM","SGD", 170],
    "ethionamide": ["SVM", "SGD", 180], # ["logReg", "GD", 0.9]
    "4-aminosalicylic_acid": ["SVM","SGD", 180],

    "capreomycin": ["SVM","SGD", 190],
    "moxifloxacin": ["SVM","SGD", 190],
    "ciprofloxacin": ["SVM","SGD", 190],
    "rifabutin": ["SVM","SGD", 100],
    "ciprofloxacin": ["SVM","SGD", 140],
    
}

clster_manual_map = {}
```

```python
sim_allDrugs_df = pd.DataFrame()
top_num = 60
training_size_percents = [.8]
simulation_iterations = 400 # 400 initially... using 60 for testing
verbose_sim = False

fpr, tpr, roc_auc = dict(), dict(), dict()
estimator_type = "SVM" #logReg or SVM
SGD_or_GD = "SGD" #SGD or GD

# kf = StratifiedKFold(n_splits=2, shuffle=False)

antibiotic_clusters = []
# "isoniazid", "ethambutol", "rifampicin", "pyrazinamide",  "ethionamide", "ofloxacin", "4-aminosalicylic_acid", "MDR", "XDR"
# for drug_name in ["isoniazid", "ethambutol", "rifampicin", "pyrazinamide",  "ethionamide", "ofloxacin", "4-aminosalicylic_acid", "MDR", "XDR", "streptomycin"]:
for drug_name in ["isoniazid", "ethambutol", "rifampicin"]:
    
    if drug_name == "MDR" or drug_name == "XDR":
        X1, y1 = get_MDR_training(resistance_data, data_to_plot, drug_name)
    else:
        y1, X1 = get_target_data(data_to_plot, resistance_data, drug_name)
        X1 = remove_key_antibiotic_clusters(X1, drug_name)
    
    print "Before removing low variance: ", X1.shape
    selector = VarianceThreshold(threshold=.01)
    selector.fit_transform(X1)
    X1 = X1[X1.columns[selector.get_support()]].copy()
    print "After removing low variance: ", X1.shape
    
    # retrieve drug machine learning parameters from dict
    estimator_type = threshold_value_dict[drug_name][0]
    SGD_or_GD = threshold_value_dict[drug_name][1]
    threshold_val = threshold_value_dict[drug_name][2]
    
    sim_all_df = pd.DataFrame()
    
    for t in tqdm(range(simulation_iterations)):
        if estimator_type == "SVM":
            if SGD_or_GD == "SGD":
                clf = SGDClassifier(loss="hinge", penalty="l1", shuffle=True, class_weight="balanced" , alpha=0.0001) 
            elif SGD_or_GD == "GD":
                clf = LinearSVC(penalty = "l1", loss="squared_hinge", dual=False, class_weight="balanced")
        elif estimator_type == "logReg":
            if SGD_or_GD == "SGD":
                clf = SGDClassifier(loss="log", penalty="l1", shuffle=True, class_weight="balanced" , alpha=0.0001) 
            elif SGD_or_GD == "GD":
                clf = LogisticRegression(penalty = "l1", class_weight="balanced")   
    
                    
        percent_of_train = random.choice(training_size_percents)
        train_x, X_test, train_y, y_test = train_test_split(X1, y1, train_size = percent_of_train, stratify=None)

        # feature selection w/ parameter "threshold"
        feat_selection = SelectFromModel(clf, threshold=threshold_val)

        # Setup pipeline for SGD-SVM feature selection
        model = Pipeline([
                  ('fs', feat_selection),  # SelectFromModel using SGD-SVM
                  ('clf', clf), 
                ])
        model.fit(train_x, train_y)

        sim_t_df = pd.DataFrame([model.named_steps["clf"].coef_[0]], 
                                columns=X1.columns[model.named_steps["fs"].get_support()])
        
        model_score_on_test = model.score(X_test, y_test)
        
        if verbose_sim == True:
            print "",
            print "# of features in simulation "+str(t)+ ":"+ str(len(sim_t_df.columns)),
            print "#R: " + str(train_y.sum()) + "#S: " + str(len(train_y.index)-train_y.sum()),
            print "% of samples: " + str(percent_of_train),
            print "Prediction Score: ", model_score_on_test
        
        # if model_score_on_test > score_threshold_test:
        sim_all_df = pd.concat([sim_all_df, sim_t_df], ignore_index=True)
        
        # y_score = model.predict(X_test)
        y_score = model.decision_function(X_test)

        fpr[t], tpr[t], _ = roc_curve(y_test.values, y_score)
        roc_auc[t] = auc(fpr[t], tpr[t])

        
    print "# of cluster features...", len(list(set(sim_all_df.columns)))

    sim_all_df_T = sim_all_df.transpose().copy()
    sim_all_df_T["svm_weight_sum"] = sim_all_df_T.apply(lambda x: abs(x).sum(), axis=1)
    sim_all_df_T_top = sim_all_df_T.sort_values("svm_weight_sum", ascending=False)[:top_num]
    
    sim_allDrugs_df = pd.concat([sim_allDrugs_df, sim_all_df[sim_all_df_T_top.index]])

    amr_clsts = []
    for x in list(sim_all_df_T_top.index):
        amr_clsts.append(str(x).split("_")[0].replace("ter", "ter "))

    amr_clsts_set = list(set(amr_clsts))
    antibiotic_clusters = antibiotic_clusters+amr_clsts_set
    
    clster_var_mapping_prod, clster_var_mapping_Rv = {}, {}
    for i in sim_all_df_T.index:
        var_index = i.split("_")[1]
        # print i
        # clster_var_mapping_prod.update({i: clust_to_prod[str(i).split("_")[0].replace("ter", "ter ")]+var_index})

        try:
            clster_var_mapping_prod.update({i: i+clust_to_prod[str(i).split("_")[0].replace("ter", "ter ")]+"_"+var_index})
            # clster_var_mapping_Rv.update({i: clust_to_rv[str(i).split("_")[0].replace("ter", "ter ")]+var_index})
        except:
            clster_var_mapping_prod.update({i: i+clust_to_prod[str(i).split("_")[0].replace("ter", "ter ")]+"_"+var_index})
            # clster_var_mapping_Rv.update({i: clust_to_prod[str(i).split("_")[0].replace("ter", "ter ")]+var_index})
    
    
    clster_to_IDENTIFIER, rv_list = clster_identifier_func(clster_manual_map, sim_all_df_T, 
                                                       clust_to_rv, rv_to_name, clust_to_prod)
    
    save_ROC_curve_figure(drug_name, fpr, tpr, roc_auc, simulation_iterations, estimator_type,SUPP_SVM_PLANES )
    
    sim_all_df_Prod, allele_to_colormap, clust_alleles = svm_top_weights_colormap(sim_all_df_T, clster_to_IDENTIFIER, 
                                                                              top_num, X1, y1, 
                                                                              clster_var_mapping_prod)
    
    sim_all_df_Prod.to_csv(SUPP_SVM_DATA+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_top_clusts.csv")
    sim_all_df_T.sort_values(["svm_weight_sum"], ascending=False)[:100].to_csv(SUPP_SVM_DATA+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_top_100alleles.csv")
    
    sim_corr_df = sim_all_df_Prod.fillna(0).T.corr().copy()
    sim_corr_df = sim_corr_df[abs(sim_corr_df)>.05].fillna(0)

    g_plot = plot_svm_heatmap(sim_corr_df, (15, 15), allele_to_colormap, "seismic") # "seismic"
    g_plot.savefig(SUPP_SVM_PLANES+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_top"+str(top_num)+"_correlation.png")
    
    sim_corr_df.to_csv(SUPP_SVM_DATA+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_top"+str(top_num)+"_correlation_matrix.csv")

    g = sns.clustermap(sim_all_df_Prod.fillna(0), 
                       col_cluster=False, row_cluster=True, cmap="seismic",
                      figsize=(18,14), xticklabels=False)
    plt.setp(g.ax_heatmap.yaxis.get_majorticklabels(), rotation=0);
    plt.setp(g.ax_heatmap.yaxis.get_majorticklabels(), rotation=0);
    g.savefig(SUPP_SVM_PLANES+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_iterations.png")
    
#    general_cluster_alleles = list(set([clst_al.split("_")[0].replace("ter", "ter ") for clst_al in clust_alleles]))

#    GEMPro_final = get_GEMPRO_df(general_cluster_alleles, clust_to_rv, rv_to_name, clster_manual_map, clust_variant_df)
#    GEMPro_final.to_csv("GEMPro_seq_dataframes/"+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_"+str(top_num)+"_SEQdf.csv")
```
