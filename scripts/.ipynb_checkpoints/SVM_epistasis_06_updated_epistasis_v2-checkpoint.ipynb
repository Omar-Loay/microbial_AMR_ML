```python
def make_table_figure(clust1, clust2, drg_name, two_cluster_df, two_clusterPercent_df, two_varPval_df, two_varOdds_df, clust_to_rv, rv_to_name, clster_manual_map, which_df_2_color, color_bar_or_not):
    """
    Automated process for building tables for the allele co-occurences.
    """
    rv_name1, rv_name2 = clust_to_rv[clust1.strip("_").replace("ter", "ter ")], clust_to_rv[clust2.strip("_").replace("ter", "ter ")]
    name1, name2 = rv_to_name[rv_name1], rv_to_name[rv_name2]

    if name1 == None and rv_name1 in clster_manual_map.keys():
        name1 = clster_manual_map[rv_name1]
    elif name1 == None:
        name1 = rv_name1

    if name2 == None and rv_name2 in clster_manual_map.keys():
        name2 = clster_manual_map[rv_name2]
    elif name2 == None:
        name2 = rv_name2
        
    # ------ Make styled dataframe by changing the index and column names, moving shit around, etc.... ------
    if which_df_2_color == "percent":
        save_occurance_df = two_clusterPercent_df.copy()
        vmin_bot=0.2
        vmax_top=0.95
    elif which_df_2_color == "pVal":
        save_occurance_df = two_varPval_df.copy()
        save_occurance_df = -np.log(save_occurance_df).replace([np.inf, -np.inf], np.nan)
        vmin_bot=0
        vmax_top=30
    elif which_df_2_color == "oddsRatio":
        save_occurance_df = two_varOdds_df.copy()
        abs_max_bound = max(abs(save_occurance_df.fillna(0).values.min()), save_occurance_df.fillna(0).values.max())
        vmin_bot = -abs_max_bound
        vmax_top = abs_max_bound

    new_cols = [str(x.split("_")[1]) if len(str(x.split("_")[1]))>0 else x for x in save_occurance_df.columns]
    new_inds = [str(x.split("_")[1]) if len(str(x.split("_")[1]))>0 else x for x in save_occurance_df.index]

    new_cols_Name = [str(x.split("_")[1]) if len(str(x.split("_")[1]))>0 else rv_name2 for x in save_occurance_df.columns]
    new_inds_Name = [str(x.split("_")[1]) if len(str(x.split("_")[1]))>0 else rv_name1 for x in save_occurance_df.index]

    allele_cols = [x for x in new_cols if "Cluster" not in x]
    # cluster_col = [x for x in new_cols if "Cluster" in x]
    cluster_col = [rv_name2]

    allele_ind = [x for x in new_inds if "Cluster" not in x]
    # cluster_ind = [x for x in new_inds if "Cluster" in x]
    cluster_ind = [rv_name1]

    # save_occurance_df.columns = new_cols
    # save_occurance_df.index = new_inds
    save_occurance_df.columns = new_cols_Name
    save_occurance_df.index = new_inds_Name

    save_occurance_df = save_occurance_df[allele_cols+cluster_col]
    save_occurance_df = save_occurance_df.reindex(index = allele_ind+cluster_ind)

    labeled_two_cluster_df = two_cluster_df.copy()
    # labeled_two_cluster_df.columns = new_cols
    # labeled_two_cluster_df.index = new_inds
    labeled_two_cluster_df.columns = new_cols_Name
    labeled_two_cluster_df.index = new_inds_Name

    labeled_two_cluster_df = labeled_two_cluster_df[allele_cols+cluster_col]
    labeled_two_cluster_df = labeled_two_cluster_df.reindex(index = allele_ind+cluster_ind)

    # ---- Add the two columns and rows to both ends of "save_occurance_df"and "labeled_two_cluster_df"
    # --- Add the two columns to right side of dataframe ----
    
    resist_vector_name = "#R"
    total_vector_name = "Total"
    
    labeled_two_cluster_df[resist_vector_name] = labeled_two_cluster_df.fillna("-")[cluster_col[0]].apply(lambda x: x.split(":")[0].strip("()").split("/")[0] if "/" in x else x)
    labeled_two_cluster_df[total_vector_name] = labeled_two_cluster_df.fillna("-")[cluster_col[0]].apply(lambda x: x.split(":")[0].strip("()").split("/")[1] if "/" in x else x)
    
    # --- Add the two rows to the bottom of dataframe ----
    row_resistant = labeled_two_cluster_df.fillna("-").loc[cluster_ind[0], :].apply(lambda x: x.split(":")[0].strip("()").split("/")[0] if "/" in x else x)
    row_resistant.name = resist_vector_name
    row_total = labeled_two_cluster_df.fillna("-").loc[cluster_ind[0], :].apply(lambda x: x.split(":")[0].strip("()").split("/")[1] if "/" in x else x)
    row_total.name = total_vector_name
    labeled_two_cluster_df = labeled_two_cluster_df.append(row_resistant)
    labeled_two_cluster_df = labeled_two_cluster_df.append(row_total)

    # --- Drop Gene ID row ---
    labeled_two_cluster_df.drop(cluster_col, axis=1, inplace=True)
    labeled_two_cluster_df.drop(cluster_ind, axis=0, inplace=True)
    
    # ---- Change save_occurance_df -----
    save_occurance_df[resist_vector_name] = save_occurance_df[cluster_col[0]]
    save_occurance_df[total_vector_name] = save_occurance_df[cluster_col[0]]
    row_occurTotal_add = save_occurance_df.loc[cluster_ind[0], :].copy()
    row_occurTotal_add.name = total_vector_name
    row_occurR_add = save_occurance_df.loc[cluster_ind[0], :].copy()
    row_occurR_add.name = resist_vector_name
    save_occurance_df = save_occurance_df.append(row_occurR_add)
    save_occurance_df = save_occurance_df.append(row_occurTotal_add)
    save_occurance_df.drop([cluster_col[0]], axis=1, inplace=True)
    save_occurance_df.drop([cluster_ind[0]], inplace=True)

    # ------ Make seaborn heatmap from the two dataframes that looks like a table ------
    # cm = sns.diverging_palette(220, 10, sep=80, n=7, l=60, as_cmap=True)
    # cm = sns.diverging_palette(220, 10, sep=1, n=7, as_cmap=True)
    # cm = sns.light_palette("red",as_cmap=True, n_colors=5)
    # cm = sns.choose_light_palette(input='husl', as_cmap=True)
    cm = sns.diverging_palette(236, 15, s=86, l=65, sep=10, n=9, as_cmap=True)
    
    # ------ reorder the columns and index based on the numeric value instead of string ------
    ordered_cols = [int(x) if x.isdigit() else x for x in save_occurance_df.columns]
    ordered_cols.sort()
    ordered_cols_string = [str(x) for x in ordered_cols]
    ordered_inds = [int(x) if x.isdigit() else x for x in save_occurance_df.index]
    ordered_inds.sort()
    ordered_inds_string = [str(x) for x in ordered_inds]

    save_occurance_df = save_occurance_df[ordered_cols_string]
    save_occurance_df.reindex(ordered_inds_string)
    
    labeled_two_cluster_df = labeled_two_cluster_df[ordered_cols_string]
    labeled_two_cluster_df.reindex(ordered_inds_string)
    
    labels = np.array(labeled_two_cluster_df.fillna("-").astype(str).copy())
    only_frac_labels = []
    for x in labels:
        row_labels = [y.split(":")[0].strip("()") for y in x]
        row_labels = ['$\\dfrac{{{0}}}{{{1}}}$'.format(y.split("/")[0], y.split("/")[1]) if "/" in y else y for y in row_labels ]
        only_frac_labels.append(row_labels)

    # plt.figure(figsize=(len(allele_ind)+1,len(allele_cols)+1))
    # plt.tight_layout()
    # rowCol_ratio_in_figure = len(save_occurance_df.index)/float(len(save_occurance_df.columns))
    if color_bar_or_not == True:
        plt.figure(figsize=(len(save_occurance_df.columns)*.65, len(save_occurance_df.index)*.5))
    else:
        plt.figure(figsize=(len(save_occurance_df.columns)*.5, len(save_occurance_df.index)*.5))
    g = sns.heatmap(save_occurance_df, annot=np.array(only_frac_labels),fmt = '',
                    cmap=cm, vmin=vmin_bot, vmax=vmax_top,
                    cbar=True, linewidths=0.6, linecolor="black", 
                    annot_kws={"color":"black"}, 
                    # annot_kws={"size": rowCol_ratio_in_figure*3, "color":"black"},
                    square=True);
                    # 
    g.set_xlabel(name1+" alleles", rotation=0); # , size=16
    g.set_ylabel(name2+"\nalleles", rotation=0); # , size=16
    g.xaxis.tick_top()
    g.xaxis.set_label_position('top') 
    g.tick_params(axis="both", which="major", length=0)
    plt.setp(g.yaxis.get_majorticklabels(), rotation=0); # , size=10
    plt.setp(g.xaxis.get_majorticklabels(), rotation=0); # , size=10
    plt.savefig("cooccurance_table_figures/"+drg_name+"_"+name1+"__"+name2+"_"+which_df_2_color+"_occurance.png",
               bbox_inches="tight", dpi=200)
    plt.close()
    return save_occurance_df, labeled_two_cluster_df


def allele_cooccurence_oddsRatio(all_1, all_2, y_drg, X_all, add_small_value_to_table, log_or_not, single_or_double_tuple):
    """
    Takes in two alleles, creates a cooccurence column, and computes the association of that
    column with the AMR phenotype.
    """
    # print X_all.shape, all_1, all_2, X_all.columns
    two_allele_df = X_all.loc[:, [all_1,all_2]].copy()
    
    if single_or_double_tuple[0] == "double":
        two_allele_df["cooccurence"] = two_allele_df[all_1] + two_allele_df[all_2]
        two_allele_df["cooccurence"][two_allele_df["cooccurence"]<2]=0
        two_allele_df["cooccurence"][two_allele_df["cooccurence"]==2]=1
    elif single_or_double_tuple[0] == "single" and all_1 == all_2:
        two_allele_df = X_all.loc[:, [all_2]].copy()
        two_allele_df["cooccurence"] = two_allele_df[single_or_double_tuple[1]]
    elif single_or_double_tuple[0] == "single":
        two_allele_df["cooccurence"] = two_allele_df[single_or_double_tuple[1]]
    
    two_allele_df["resistant"] = y_drg
    
    cooccurence_resistant = float(len(two_allele_df[(two_allele_df["cooccurence"]==1) & (two_allele_df["resistant"]==1)].index))
    cooccurence_susceptible = float(len(two_allele_df[(two_allele_df["cooccurence"]==1) & (two_allele_df["resistant"]==0)].index))
    
    no_cooccurence_resistant = float(len(two_allele_df[(two_allele_df["cooccurence"]==0) & (two_allele_df["resistant"]==1)].index))
    no_cooccurence_susceptible = float(len(two_allele_df[(two_allele_df["cooccurence"]==0) & (two_allele_df["resistant"]==0)].index))
    
    # add_small_value_to_table = .1
    # if one of the cells is 0, there won't be an odds ratio. To get past this, we add a small value of .5 to each value
    if cooccurence_resistant==0 or cooccurence_susceptible==0 or no_cooccurence_resistant==0 or no_cooccurence_susceptible==0:
        cooccurence_resistant+=add_small_value_to_table
        cooccurence_susceptible+=add_small_value_to_table
        no_cooccurence_resistant+=add_small_value_to_table
        no_cooccurence_susceptible+=add_small_value_to_table
    
    odds_ratio = (cooccurence_resistant/cooccurence_susceptible)/(no_cooccurence_resistant/no_cooccurence_susceptible)
    
    if log_or_not == True:
        odds_ratio = np.log(odds_ratio)
    
    return odds_ratio


def allele_cooccurence_oddsRatio_xNOTy(all_1, all_2, y_drg, X_all, add_small_value_to_table, log_or_not):
    """
    Takes in two alleles, creates a cooccurence column, and computes the association of that
    column with the AMR phenotype.
    """
    # print X_all.shape, all_1, all_2, X_all.columns
    two_allele_df = X_all.loc[:, [all_1, all_2]].copy()
    two_allele_df["resistant"] = y_drg
    
    cooccurence_resistant_x1_NOT_y1 = float(len(two_allele_df[(two_allele_df[all_1]==1) & (two_allele_df[all_2]==0) & (two_allele_df["resistant"]==1)].index))
    cooccurence_susceptible_x1_NOT_y1 = float(len(two_allele_df[(two_allele_df[all_1]==1) & (two_allele_df[all_2]==0) & (two_allele_df["resistant"]==0)].index))

    cooccurence_resistant_NOT_x1_NOT_y1 = float(len(two_allele_df[(two_allele_df[all_1]==0) & (two_allele_df[all_2]==0) & (two_allele_df["resistant"]==1)].index))
    cooccurence_susceptible_NOT_x1_NOT_y1 = float(len(two_allele_df[(two_allele_df[all_1]==0) & (two_allele_df[all_2]==0) & (two_allele_df["resistant"]==0)].index))

#     print "cooccurence_resistant_x1_NOT_y1--", cooccurence_resistant_x1_NOT_y1
#     print "cooccurence_susceptible_x1_NOT_y1--", cooccurence_susceptible_x1_NOT_y1     
#     print "cooccurence_resistant_NOT_x1_NOT_y1--", cooccurence_resistant_NOT_x1_NOT_y1
#     print "cooccurence_susceptible_NOT_x1_NOT_y1--", cooccurence_susceptible_NOT_x1_NOT_y1

    if cooccurence_resistant_x1_NOT_y1==0 or cooccurence_susceptible_x1_NOT_y1==0 or cooccurence_resistant_NOT_x1_NOT_y1==0 or cooccurence_susceptible_NOT_x1_NOT_y1==0:
        cooccurence_resistant_x1_NOT_y1+=add_small_value_to_table
        cooccurence_susceptible_x1_NOT_y1+=add_small_value_to_table
        cooccurence_resistant_NOT_x1_NOT_y1+=add_small_value_to_table
        cooccurence_susceptible_NOT_x1_NOT_y1+=add_small_value_to_table

    odds_ratio = (cooccurence_resistant_x1_NOT_y1/cooccurence_susceptible_x1_NOT_y1)/(cooccurence_resistant_NOT_x1_NOT_y1/cooccurence_susceptible_NOT_x1_NOT_y1)
    if log_or_not == True:
        odds_ratio = np.log(odds_ratio)
    
    return odds_ratio


def drop_collinearity_variables(X_alleles, Y_alleles, interaction_terms_list, X_logReg_df):
    
    drop_collinearity_alleles = []
    for allele_test_x in X_alleles:
        for allele_test_y in Y_alleles:
            allele_test_xy = allele_test_x+":"+allele_test_y
            if allele_test_xy in interaction_terms_list:
                strains_collin_x_xy = X_logReg_df[X_logReg_df[allele_test_x] == X_logReg_df[allele_test_xy]]
                strains_collin_y_xy = X_logReg_df[X_logReg_df[allele_test_y] == X_logReg_df[allele_test_xy]]
                strains_collin_x_y = X_logReg_df[X_logReg_df[allele_test_y] == X_logReg_df[allele_test_x]]
                if len(strains_collin_x_xy) == len(X_logReg_df.index):
                    print "collinearity...x_xy", len(strains_collin_both)
                    # drop_collinearity_alleles.append(allele_test_xy)
                    drop_collinearity_alleles.append(allele_test_x)
                    
                elif len(strains_collin_y_xy) == len(X_logReg_df.index):
                    print "collinearity...y_xy", len(strains_collin_both)
                    # drop_collinearity_alleles.append(allele_test_xy)
                    drop_collinearity_alleles.append(allele_test_y)
                    
                elif len(strains_collin_x_y) == len(X_logReg_df.index):
                    print "collinearity...x_y", len(strains_collin_both)
                    # ---- If the presence absence vectors of allele x and allele y are exactly the same,
                    # drop these alleles along with their interaction term.
                    drop_collinearity_alleles.extend([allele_test_x, allele_test_y])
    
    # ---- Check for the case where 3 interaction terms should be removed if contained by a single allele.
    
    for allele_test_x in X_alleles:
        allele_test_x_ALL_y = [str(x) for x in interaction_terms_list if str(allele_test_x)+":" in x]
        strains_collin_x_ALL_xy = X_logReg_df[X_logReg_df[allele_test_x] == X_logReg_df[allele_test_x_ALL_y].sum(axis=1)].index
        
        if len(strains_collin_x_ALL_xy) == len(X_logReg_df.index):
            # -- new drop code
            if allele_test_x not in drop_collinearity_alleles:
                drop_collinearity_alleles.append(allele_test_x)
                
#             for z in allele_test_x_ALL_y:
#                 if z not in drop_collinearity_alleles:
#                     drop_collinearity_alleles.append(z)

    for allele_test_y in Y_alleles:
        allele_test_y_ALL_x = [str(x) for x in interaction_terms_list if ":"+str(allele_test_y) in x]
        strains_collin_y_ALL_xy = X_logReg_df[X_logReg_df[allele_test_y] == X_logReg_df[allele_test_y_ALL_x].sum(axis=1)].index
        
        if len(strains_collin_y_ALL_xy) == len(X_logReg_df.index):
            # -- new drop code
            if allele_test_y not in drop_collinearity_alleles:
                drop_collinearity_alleles.append(allele_test_y)
            
#             for z in allele_test_y_ALL_x:
#                 if z not in drop_collinearity_alleles:
#                     drop_collinearity_alleles.append(z)

    print drop_collinearity_alleles
    
    return drop_collinearity_alleles

# drop_co_alleles = drop_collinearity_variables(X_alleles, Y_alleles, interaction_terms_list, X_logReg_df)
```

```python
top_allele_pairs = 60 # 100, to 50, (50- total of 50 gene-gene)
top_allele_weighted = 30 # 30, (15- total of 50 gene-gene)
correlation_allele_count = 60 # 100, to 50

drug_sig_gene_interact_dict = {}
drug_sig_cluster_interact_dict = {}

# for drug_name in ["isoniazid","ethambutol", "rifampicin", "pyrazinamide", "ethionamide", "ofloxacin", "4-aminosalicylic_acid","streptomycin", "MDR","XDR"]: # , 
for drug_name in ["ethambutol"]:

    print "-------------", drug_name, "-------------" 
    
    estimator_type, SGD_or_GD = seqDF_dict[drug_name]
    drug_allele_correlation_df = pd.read_csv(
        "SVM_dataframes/"+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_correlation_matrix.csv",  # top40_
                                        index_col=0)
    drug_allele_topWeights_df = pd.read_csv(
        "SVM_dataframes/"+drug_name+"_"+estimator_type+"_"+SGD_or_GD+"_top_clusts.csv", 
                                        index_col=0)
    top_weighted_alleles = list(drug_allele_topWeights_df.index[:top_allele_weighted])
    
    sig_gene_interact_list = []
    
    drug_sig_gene_interact_dict[drug_name] = {}
    drug_sig_cluster_interact_dict[drug_name] = {}
    
    # put dataframe in form to look at top correlations
    sim_corr_df = drug_allele_correlation_df.copy()
    abs_corr_df = sim_corr_df.abs()
    os = (abs_corr_df.where(np.triu(np.ones(abs_corr_df.shape), k=1).astype(np.bool))
                 .stack()
                 .sort_values(ascending=False))

    count = 0
    for (allele_1, allele_2) in os.index[:top_allele_pairs]:
        var1, var2 = allele_1.split("_")[-1], allele_2.split("_")[-1]
        if count < correlation_allele_count:
            try:

                clst_1,prod_1, clst_2,prod_2 = allele_1.split("Rv")[0],allele_1.split(", ")[1], allele_2.split("Rv")[0],allele_2.split(", ")[1]
                rv_1, rv_2= "Rv"+allele_1.split("Rv")[1].split(", ")[0], "Rv"+allele_2.split("Rv")[1].split(", ")[0]

                if clst_1 != clst_2:
                    if (rv_1, rv_2) not in sig_gene_interact_list and (rv_2, rv_1) not in sig_gene_interact_list:
                        # if os.loc[(allele_1, allele_2)] > 0.30:
                        if allele_1 in top_weighted_alleles or allele_1 in top_weighted_alleles:
                            sig_gene_interact_list.append((rv_1, rv_2))
                            drug_sig_gene_interact_dict[drug_name].update({(rv_1, rv_2): os.loc[(allele_1, allele_2)]})
                            drug_sig_cluster_interact_dict[drug_name].update({(clst_1, clst_2): os.loc[(allele_1, allele_2)]})
                            # print prod_1, "||| ",prod_2
                            # print rv_1, rv_2
                            # print (clst_1, clst_2), os.loc[(allele_1, allele_2)]
                            # print var1, var2
                            # print 
                            count = count + 1
                    
            except:
                print "NO GOOD: ", allele_1, allele_2
                

pairwise_table_groups = drug_sig_cluster_interact_dict
clust_alleles = ["what"]
color_measure = "oddsRatio" # "percent", "oddsRatio", "pVal"
pVal_test_type = "chi2" # or f_classif - only useful if color_measure = "pVal"
add_zeroCell = 0.5 # Adds this value to each cell in the odds ratio table. Happens when there is a 0
pVal_threshold_epistasis = 0.05

antibiotic_gene_gene_interactions = {}

for antibiotic_name, cluster_groups in pairwise_table_groups.iteritems():
    
    antibiotic_gene_gene_interactions[antibiotic_name] = {}
    
    row_index_location = 3
    
    # ", "ethionamide", "ofloxacin", "amikacin" "4-aminosalicylic_acid"
    if antibiotic_name == "MDR" or antibiotic_name == "XDR":
        X1, y1 = get_MDR_training(resistance_data, data_to_plot, antibiotic_name)
    else:
        y1, X1 = get_target_data(data_to_plot, resistance_data, antibiotic_name)
        X1 = remove_key_antibiotic_clusters(X1, antibiotic_name)
        
    
    for (clust1, clust2) in cluster_groups.keys():
        
        rv_id_1, rv_id_2 = clust_to_rv[clust1], clust_to_rv[clust2]
        clust1 = clust1.replace("ter ", "ter")+"_"
        clust2 = clust2.replace("ter ", "ter")+"_"
        # write code for specific case
        
        two_clust_df, two_clustPercent_df = get_2allele_resist_dataframes(clust1, clust2, antibiotic_name, "All",
                                                                              y1,clust_variant_df.loc[y1.index, :], 
                                                                              clust_variant_df, clust_alleles, False)
        if color_measure == "oddsRatio":
            two_alleleOdds_df = return_2allele_oddsRatio_df(two_clustPercent_df, two_clust_df, 
                                           clust_variant_df.loc[y1.index, :], y1, add_zeroCell, log_or_no=True)
            two_allelePval_df = pd.DataFrame()
        
        # note that the last input is either pVal or percent depending on how to color the cell.
        if antibiotic_name == '4-aminosalicylic_acid':
            antibiotic_name = "PAS"
        save_occur_df, labeled_2cluster_df = make_table_figure(clust1, clust2, antibiotic_name, two_clust_df, 
                                                               two_clustPercent_df, 
                                                               two_allelePval_df, two_alleleOdds_df,
                                                               clust_to_rv, 
                                                               rv_to_name, clster_manual_map, color_measure,
                                                              color_bar_or_not = True)
        
        X_alleles = [x for x in two_alleleOdds_df.index if x[-1]!="_"]
        Y_alleles = [x for x in two_alleleOdds_df.columns if x[-1]!="_"]
        
        X1 = X1.fillna(0)
        two_gene_interact_alleles = X1.loc[:,X_alleles+Y_alleles].copy()
        # print two_gene_interact_alleles.shape
        # print 

        # ---- start loop here.... -----
        interaction_terms_list = []
        for allele_x in X_alleles:
            for allele_y in Y_alleles:
                if two_alleleOdds_df.loc[allele_x, allele_y]!=0:
                    interaction_terms_list.append(allele_x+":"+allele_y)
                    
                    # --- Write allele-allele logistic regression: 
                    # ---- Y ~ a + a1*x1 + a2*x2 + a3*x1*x2

                    logReg_string_equation = allele_x + " + " + allele_y + " + " + allele_x+":"+allele_y


        logReg_string_equation = logReg_string_equation.strip(" + ")
        logReg_string_equation = "AMR_phenotype ~ "+logReg_string_equation
        # print logReg_string_equation

        two_gene_interact_alleles["AMR_phenotype"] = y1

        # Use patsy package to turn this string equation into something usable by python.
        y_logReg, X_logReg = patsy.dmatrices(logReg_string_equation, two_gene_interact_alleles)

        # print 
        # print "single:", len(two_gene_interact_alleles.columns), "| interactions:", len(interaction_terms_list)
        
        y_logReg_df = pd.DataFrame(np.asarray(y_logReg), index = y1.index, columns=["AMR_phenotype"])
        X_logReg_df = pd.DataFrame(np.asarray(X_logReg), index = X1.index, 
                                   columns=X_logReg.design_info.column_names)
        X_logReg_df.drop(["Intercept"], axis=1, inplace=True)
        
        # ----- GET RID OF MAJOR COLLINEARITY -----
        drop_co_alleles = drop_collinearity_variables(X_alleles, Y_alleles, interaction_terms_list, X_logReg_df)
        # print X_logReg_df.shape
        X_logReg_df.drop(drop_co_alleles, axis=1, inplace=True)
        # print X_logReg_df.shape
        
        try:
            logit = sm.Logit(y_logReg_df["AMR_phenotype"], X_logReg_df)
            result = logit.fit(maxiter=100, method='bfgs',disp=0)

            # print list(result.pvalues[result.pvalues<0.005].index)
            # pVal_thresh_series = result.pvalues[result.pvalues<pVal_threshold_epistasis]
            pVal_thresh_series = result.pvalues
            pVal_thresh_df = pd.DataFrame(pVal_thresh_series)
            pVal_thresh_df.columns = ["pVal"]
            pVal_thresh_df["aic"], pVal_thresh_df["bic"], pVal_thresh_df["pseudo_r2"] = result.aic, result.bic, result.prsquared
            antibiotic_gene_gene_interactions[antibiotic_name][(clust1, clust2)] = dict(pVal_thresh_df.T)
            # antibiotic_gene_gene_interactions[antibiotic_name][(clust1, clust2)] = dict(result.pvalues[result.pvalues<pVal_threshold_epistasis])
        except:
            continue
            # print "optimization error... likely collinearity..."
```
