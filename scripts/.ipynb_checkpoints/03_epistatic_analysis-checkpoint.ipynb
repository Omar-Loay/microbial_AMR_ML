## 04.B 
### Determine which of the potential gene-gene interactions are significant using logistic regression models.
1. Compute all significant gene-gene interactions.
2. Create a co-occurrance table for each gene-gene interaction

Parameters
- **color_measure**  - Metric used to color the allele co-occurrance table (oddsRatio, percent, oddsRatio", pVal). $\fbox{Default is oddsRatio}$
- **pVal_test_type** - Type of significance test (chi2, f_classif). Note that f_classif is only useful if color_measure = "pVal". $\fbox{Default is chi2}$
- **add_zeroCell** - Adds this value to each cell in the odds ratio table in order to account for the cases where there is a 0. $\fbox{Default is 0.5}$

```python
# set parameters
color_measure = "oddsRatio" 
pVal_test_type = "chi2" 
add_zeroCell = 0.5 
pVal_threshold_epistasis = 0.05
```

Since each optimization inside the profile likelihood generates a warning (due to the random slope variance being close to zero), we turn off the warnings here.

```python
def drop_collinearity_variables(X_alleles, Y_alleles, interaction_terms_list, X_logReg_df, verbose):
    """
    Filter out allelic variables in epistatic logistic regression model according to the
    following criteria.
        Criteria: 
            - x == xy, drop allele xy
            - 
    """
    drop_collinearity_alleles = []
    for allele_test_x in X_alleles:
        for allele_test_y in Y_alleles:
            allele_test_xy = allele_test_x+":"+allele_test_y
            if allele_test_xy in interaction_terms_list:
                strains_collin_x_xy = X_logReg_df[X_logReg_df[allele_test_x] == X_logReg_df[allele_test_xy]]
                strains_collin_y_xy = X_logReg_df[X_logReg_df[allele_test_y] == X_logReg_df[allele_test_xy]]
                strains_collin_x_y = X_logReg_df[X_logReg_df[allele_test_y] == X_logReg_df[allele_test_x]]
                if len(strains_collin_x_xy) == len(X_logReg_df.index):
                    if verbose==True:
                        print "collinearity...x_xy ... dropping: ", allele_test_x, len(strains_collin_x_xy)
                    
                    drop_collinearity_alleles.append(allele_test_xy) # off before.... # got oxcA
                    # drop_collinearity_alleles.append(allele_test_x) # original
                    
                elif len(strains_collin_y_xy) == len(X_logReg_df.index):
                    if verbose==True:
                        print "collinearity...y_xy ... dropping: ", allele_test_y, len(strains_collin_x_xy)
                    drop_collinearity_alleles.append(allele_test_xy) # off before.... # got oxcA
                    # drop_collinearity_alleles.append(allele_test_y) # original
                    
                elif len(strains_collin_x_y) == len(X_logReg_df.index):
                    if verbose==True:
                        print "collinearity...x_y ... dropping: ", allele_test_x, allele_test_y, allele_test_xy, len(strains_collin_x_xy)
                    # ---- If the presence absence vectors of allele x and allele y are exactly the same,
                    # drop these alleles along with their interaction term.
                    drop_collinearity_alleles.extend([allele_test_x, allele_test_y, allele_test_xy]) # , allele_test_xy
    
    # ---- Check for the case where 3 interaction terms should be removed if contained by a single allele.
    
    for allele_test_x in X_alleles:
        allele_test_x_ALL_y = [str(x) for x in interaction_terms_list if str(allele_test_x)+":" in x]
        strains_collin_x_ALL_xy = X_logReg_df[X_logReg_df[allele_test_x] == X_logReg_df[allele_test_x_ALL_y].sum(axis=1)].index
        
        if len(strains_collin_x_ALL_xy) == len(X_logReg_df.index):
            # -- new drop code... remove allele
            if allele_test_x not in drop_collinearity_alleles:
                if verbose==True:
                    print "collinearity...x_ALL_y ... dropping: ", allele_test_x
                drop_collinearity_alleles.append(allele_test_x) # original
                
#             for z in allele_test_x_ALL_y:
#                 if z not in drop_collinearity_alleles:
#                     drop_collinearity_alleles.append(z)

    for allele_test_y in Y_alleles:
        allele_test_y_ALL_x = [str(x) for x in interaction_terms_list if ":"+str(allele_test_y) in x]
        strains_collin_y_ALL_xy = X_logReg_df[X_logReg_df[allele_test_y] == X_logReg_df[allele_test_y_ALL_x].sum(axis=1)].index
        
        if len(strains_collin_y_ALL_xy) == len(X_logReg_df.index):
            # -- new drop code
            if allele_test_y not in drop_collinearity_alleles:
                if verbose==True:
                    print "collinearity...y_ALL_x ... dropping: ", allele_test_y
                drop_collinearity_alleles.append(allele_test_y) # original
            
#             for z in allele_test_y_ALL_x:
#                 if z not in drop_collinearity_alleles:
#                     drop_collinearity_alleles.append(z)
    if verbose==True:
        print "Dropped variables in locus-locus logistic model: ", drop_collinearity_alleles
    
    return drop_collinearity_alleles
```
